```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_spaceweather(csv_path="../data/space_weather_aligned.csv"):
    df = pd.read_csv(csv_path, parse_dates=["time"])
    df = df.set_index("time").sort_index()

    fig, axes = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

    df["BZ"].plot(ax=axes[0], title="IMF BZ (nT)")
    df["V_SW"].plot(ax=axes[1], title="Solar Wind Speed (km/s)")
    df["Dst"].plot(ax=axes[2], title="Dst Index (nT)")
    df["Kp"].plot(ax=axes[3], title="Kp Index")

    plt.tight_layout()
    plt.show()

plot_spaceweather()
```
